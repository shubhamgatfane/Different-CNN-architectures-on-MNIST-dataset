```python
from __future__ import print_function
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, BatchNormalization
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K

batch_size = 128
num_classes = 10
epochs = 12

# input image dimensions
img_rows, img_cols = 28, 28

# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()

if K.image_data_format() == 'channels_first':
    x_train = x_train.reshape(x_train.shape[0], 1, img_rows, img_cols)
    x_test = x_test.reshape(x_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
    x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

# Initialising the CNN model
model = Sequential()

# 1st Convolution Layer #Pooling
model.add(Conv2D(32, kernel_size=(2, 2), activation='relu', input_shape=input_shape,padding='same'))
model.add(BatchNormalization())


# Adding 2nd convolutional layer #Pooling
model.add(Conv2D(32, (2, 2), activation='relu',padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2),strides=(2, 2)))     


# Adding 3rd convolutional layer
model.add(Conv2D(64, (2, 2), activation='relu',padding='same'))
model.add(BatchNormalization())


# Adding 4th convolutional layer
model.add(Conv2D(64, (2, 2), activation='relu',padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2),strides=(2, 2)))

# Adding 5th convolutional layer
model.add(Conv2D(128, (2, 2), activation='relu',padding='same'))
model.add(BatchNormalization())

# Adding 6th convolutional layer
model.add(Conv2D(128, (2, 2), activation='relu',padding='same'))
model.add(BatchNormalization())

# Adding 7th convolutional layer
model.add(Conv2D(256, (2, 2), activation='relu',padding='same'))
model.add(BatchNormalization())


#Pooling
model.add(MaxPooling2D(pool_size=(2, 2),strides=(2, 2)))       

#Dropout
model.add(Dropout(0.30))

#Flatten Operation
model.add(Flatten())

#Applying ANN
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])

model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
